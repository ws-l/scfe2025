```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
from sklearn.preprocessing import StandardScaler
import pandas as pd
df0 = pd.read_csv("BostonHousing.csv")
scaled = StandardScaler().fit_transform( df0 )
df1 = pd.DataFrame( scaled, columns = df0.columns)
X_train, X_test, y_train, y_test = train_test_split(df1[["crim", "nox"]],
                                                    df1.medv, train_size=0.7)
model = LinearRegression()
model.fit( X_train, y_train  ) #print( model.coef_, model.intercept_ )
pred = model.predict(X_test)
root_mean_squared_error( y_test, pred)  #RMSE
model.score(X_train, y_train)   #R-squared, 0~1
```
