```python
#creditapprovaluci.csv 읽으셔서
#결측치 체크
#결측치 제거
#dtype이 숫자인 것만 고르기
#파티셔닝
#A16: target 분류모형
#dt, rf
#성능비교
import pandas as pd
df1 = pd.read_csv("creditApprovalUCI.csv")
df1.isnull().mean()
df1.dropna(inplace=True)
df2 = df1.select_dtypes(exclude="object")

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( df2.drop("A16", axis=1),
                                                    df2.A16,
                                                     stratify=df2.A16)
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

from sklearn.metrics import classification_report
print(classification_report(y_test, dt.predict(X_test)))
print(classification_report(y_test, rf.predict(X_test)))
```

```python
import pandas as pd
df1 = pd.read_csv("creditset2.csv")
df2 = df1[["age", "loan", "income"]]
#scaling, df3
from sklearn.preprocessing import MinMaxScaler
scaled = MinMaxScaler().fit_transform( df2 )
df3 = pd.DataFrame( scaled, columns=df2.columns)
from sklearn.cluster import AgglomerativeClustering, KMeans
cl = KMeans(n_clusters=10)
cl.fit(df3)
cl.labels_
df2["cluster"] = cl.labels_
df2.groupby("cluster").mean()
```

```python
from pykrx import stock
df4 = stock.get_market_ohlcv("20250123")
df5 = stock.get_market_fundamental("20250123")
df6 = df4.join(df5)
```

```python
df7 = df6[["종가", "거래량", "BPS", "PER", "PBR", "EPS", "DIV","DPS"]]
df7.dropna(inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaled = MinMaxScaler().fit_transform( df7 )
df8 = pd.DataFrame( scaled, columns=df7.columns, index=df7.index)

from sklearn.cluster import AgglomerativeClustering, KMeans
cl = KMeans(n_clusters=5)
cl.fit(df8)
cl.labels_
df7["cluster"] = cl.labels_
df7.groupby("cluster").mean()
```
